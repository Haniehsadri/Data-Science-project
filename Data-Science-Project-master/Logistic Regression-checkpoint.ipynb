In this Part I use Logistic Regression to predict whether a day is holiday or not based on Number of total vehicles and hour

```python
import numpy as np
import pandas as pd

data = pd.read_csv('Logistic Regression Data 1.csv', header=None, names=['Total Vehicles', 'From(H)', 'Is Holiday'])
data
```

```python

X = data.iloc[:,0:-1].values


maxX = np.max(X, axis=0)
minX = np.min(X, axis=0)
X = (X-minX)/(maxX-minX)




X = np.insert(X, 0, 1, axis=1)



y = data.iloc[:,-1:].values 



where_are_zeros = (y==0)
y[where_are_zeros] = -1
```

Let's visualize Data. We can create a function for that.

```python
import matplotlib.pyplot as plt
def visualize(X, y, col1=1, col2=2):
    positive_indexes = np.where(y == 1)[0]  
    negative_indexes = np.where(y == -1)[0]

    positive = X[positive_indexes]  
    negative = X[negative_indexes]  

    fig, ax = plt.subplots(figsize=(12,8))
    ax.scatter(positive[:,1:2], positive[:,2:], s=50, c='g', marker='o', label='Holiday')
    ax.scatter(negative[:,1:2], negative[:,2:], s=50, c='r', marker='x', label='Non-Holiday')
    ax.legend()
    ax.set_xlabel('Total Vehicles')
    ax.set_ylabel('Hour')
    plt.show()
    

visualize(X,y)
```

```python
n=X.shape[0]
w = np.ones((1,X.shape[1]))
def error(x,y,w):
    return np.log(1+np.exp(-y*x@w.T))
def error_mean(X,y,w):
    return sum(error(X,y,w))/n
def grad(x,y,w):
    return y*x/(1+np.exp(y*x@w.T))
def grad_mean(X,y,w):
    return sum(grad(X,y,w))/-n
def fit(X,y,kappa,iter):
    w = np.zeros((1,X.shape[1]))
    E = []
    for i in range(iter):
       w=w-kappa*grad_mean(X,y,w) 
       E.append(error_mean(X,y,w)) 
    return w,E
w,E = fit(X,y,1,1000)
plt.plot(E)
plt.show()
print(w)
```

We then use this function to score the training accuracy of our classifier.

```python
def predict(w, X):
    p=(1/(1+np.exp(-1*X@w.T)))>0.5
    y_pred=np.ones(p.shape)
    for i in range(len(p)):
        if p[i,0]==False:
            y_pred[i,0]=-1
    return y_pred
def accuracy(y,y_pred):
    accurateNumber=0
    for i in range(len(y)):
        if y[i,0]==y_pred[i,0]:
            accurateNumber=accurateNumber+1
    return accurateNumber/len(y)
y_pred = predict(w,X)
print( accuracy(y,y_pred) )
```

```python
def split_train_test(X,y,pct=80):
    n = X.shape[0]
    s = round(n * pct / 100)
    
    indices = np.random.permutation(n)
    train_idx, test_idx = indices[:s], indices[s:]
    
    X_train, X_test = X[train_idx,:], X[test_idx,:]
    y_train, y_test = y[train_idx,:], y[test_idx,:]
    
    return X_train, y_train, X_test, y_test
```

```python
X_train, y_train, X_test, y_test = split_train_test(X,y,pct=80)
w,E = fit(X_train,y_train,1,1000)
print(w)
plt.plot(E)
plt.show()
y_pred = predict(w,X_test)
print( accuracy(y_test,y_pred) )
```
