In this Part I use Linear Regression to predict Number of unauthorized speed based on Number of total vehicles and average of velocity

```python
import numpy as np
import pandas as pd

data = pd.read_csv('Linear Regression Data 1.csv', header=None, names=['Total Vehicles', 'AVG Velocity', 'No of Unauthorized Speed'])
data
```

```python

X = data.iloc[:,0:-1].values


maxX = np.max(X, axis=0)
minX = np.min(X, axis=0)
meanX=np.sum(X, axis=0)/X.shape[0]

X = (X-meanX)/(maxX-minX)



X = np.insert(X, 0, 1, axis=1)



y = data.iloc[:,-1:].values.T
```

```python
def error4(X,y,w):
    n=X.shape[0]
    return (1/(2*n))*(np.sum((y-w@X.T)**2))
```

```python
n=X.shape[0]
def grad4(x,y,w):
    return ((1/n)*(np.sum((y-w@X.T)@X,axis=0,keepdims=True)))
```

```python
import matplotlib.pyplot as plt
def fit4(X,y,kappa,iter):
    w = np.zeros((1,X.shape[1]))
    E = []

  
    for i in range(iter):
       E.append(error4(X,y,w)) 
       w=w+kappa*grad4(X,y,w) 
       
    return w,E
i=0.3
while i<2:
    w,E = fit4(X,y,i,100)
    plt.plot(E, label="Kappa = " + str(i))
    i = i + 0.4
plt.legend()
plt.show()
```
